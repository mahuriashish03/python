```python
class Animal():
    def __init__(self):
        print("Animal Create")

    def who_am_i(self):
        print("I am animal")

    def eat(self):
        print("I am eating")
    
```

```python
class Dog(Animal): 
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")

    #Method overriding
    def who_am_i(self):
        print("I am a dog")

    def bark(self):
        print("WOOF!")
```

```python
my_dog = Dog()
```

```python
my_dog.who_am_i()
```

```python
my_dog = Animal()
```

```python
my_dog.who_am_i()
```

##Polymorphism

```python
class Dog():
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name + "says woof!"
```

```python
class Cat():
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name + "says meow!"
```

```python
niko = Dog("niko")
caty = Cat("caty")
```

```python
print(niko.speak())
```

```python
print(caty.speak())
```

```python
for pet_class in [niko, caty]:
    print(type(pet_class))
    print(pet_class.speak())
```

#Real polymorphism example

```python
class Animal():
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass needs to implement this method")

class Dog(Animal):
    def speak(self):
        print(self.name + " says woof!")

class Cat(Animal):
    def speak(self):
        print(self.name + " says meow!")
```

```python
my_dog = Dog("Shiro")
my_cat = Cat("Katy")
```

```python
my_dog.speak()
```

```python
my_cat.speak()
```
