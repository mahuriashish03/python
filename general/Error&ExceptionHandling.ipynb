```python
try:
    result = 10 + 10

except:
    print("Error occured")

finally:
    print(result)
    print("I always run")
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input("Enter the number: "))
        except:
            print("Woops not a number")
            continue
        else:
            print("Yes thank you")
            break
        finally:
            print("End of try/catch/finally")
```
