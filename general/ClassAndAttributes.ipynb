```python
class Circle():
    pi = 3.14

    def __init__(self, radius=1): 
        self.radius = radius
        self.area = Circle.pi *  radius**2

    def circumference(self):
        return 2 * Circle.pi * self.radius 
```

```python
my_circle = Circle()
```

```python
my_circle.circumference()
```

```python
my_circle = Circle(30)
```

```python
my_circle.circumference()
```

```python
my_circle.area
```
